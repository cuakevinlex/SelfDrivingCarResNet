```python
class Lane_net(nn.Module):
    def __init__(self):
        super(Lane_net, self).__init__()

        # linear activation
        self.relu_layer = nn.ReLU(inplace=True)
        # max pool 
        self.pool = nn.MaxPool2d(kernel_size=4, stride=3)

        # conv2d(n_samples, channels, height, width)
        # first conv layer
        self.cnn_layer1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        
        # batch normalization
        self.batch1 = nn.BatchNorm2d(32)

        # second conv layer
        self.cnn_layer2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.batch2 = nn.BatchNorm2d(64)

        # third conv layer
        self.cnn_layer3 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.batch3 = nn.BatchNorm2d(128)
        
        self.cnn_layer4 = nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1)
        self.batch4 = nn.BatchNorm2d(128)

        self.linear_layer1 = nn.Linear(128*35, 2048)
        self.linear_layer2 = nn.Linear(2048,1024)
        self.linear_layer3 = nn.Linear(1024,18)
        

    def forward(self, x):
        # first conv layer
        x = self.cnn_layer1(x)
        x = self.batch1(x)
        x = self.relu_layer(x)
        x = self.pool(x)

        # second conv layer
        x = self.cnn_layer2(x)
        x = self.batch2(x)
        x = self.relu_layer(x)
        x = self.pool(x)

        # third conv layer
        x = self.cnn_layer3(x)
        x = self.batch3(x)
        x = self.relu_layer(x)
        x = self.pool(x)
        
        # third conv layer
        x = self.cnn_layer4(x)
        x = self.batch4(x)
        x = self.relu_layer(x)
        x = self.pool(x)

        # linear 
        x = x.view(x.size(0), -1)
        #print(x.size())
        x = self.linear_layer1(x)
        x = self.linear_layer2(x)
        x = self.linear_layer3(x)
        return x
```

```python
class Lane_drop(nn.Module):
    def __init__(self):
        super(Lane_drop, self).__init__()

        # linear activation
        self.relu_layer = nn.ReLU(inplace=True)
        # max pool 
        self.pool = nn.MaxPool2d(kernel_size=4, stride=3)

        # conv2d(n_samples, channels, height, width)
        # first conv layer
        self.cnn_layer1 = nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1)
        
        # batch normalization
        self.batch1 = nn.BatchNorm2d(32)

        # second conv layer
        self.cnn_layer2 = nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1)
        self.batch2 = nn.BatchNorm2d(64)

        # third conv layer
        self.cnn_layer3 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.batch3 = nn.BatchNorm2d(128)
        
        self.cnn_layer4 = nn.Conv2d(128, 128, kernel_size=3, stride=1, padding=1)
        self.batch4 = nn.BatchNorm2d(128)

        self.linear_layer1 = nn.Linear(128*35, 2048)
        self.linear_layer2 = nn.Linear(2048,1024)
        self.linear_layer3 = nn.Linear(1024,18)
        self.dropout = nn.Dropout(0.25)

    def forward(self, x):
        # first conv layer
        x = self.cnn_layer1(x)
        x = self.batch1(x)
        x = self.relu_layer(x)
        x = self.pool(x)

        # second conv layer
        x = self.cnn_layer2(x)
        x = self.batch2(x)
        x = self.relu_layer(x)
        x = self.pool(x)

        # third conv layer
        x = self.cnn_layer3(x)
        x = self.batch3(x)
        x = self.relu_layer(x)
        x = self.pool(x)
        
        # third conv layer
        x = self.cnn_layer4(x)
        x = self.batch4(x)
        x = self.relu_layer(x)
        x = self.pool(x)

        # linear 
        x = x.view(x.size(0), -1)
        #print(x.size())
        x = self.linear_layer1(x)
        x = self.linear_layer2(x)
        x = self.linear_layer3(x)
        x = self.dropout(x)
        
        return x
```

```python
# Load the model based on DenseNet-161
# lane_net = torchvision.models.mobilenet_v2(pretrained=True)
lane_net = torchvision.models.densenet121(pretrained=True)

# # freeze the layers
# for param in lane_net.parameters():
#    param.requires_grad = False

# Modify the last layer for mobil
# num_ftrs = lane_net.classifier[1].in_features
# lane_net.classifier[1] = nn.Linear(num_ftrs, 18)

# modify for dense
num_ftrs = lane_net.classifier.in_features
lane_net.classifier = nn.Linear(num_ftrs, 18)


# lane net
# lane_net = Lane_net()

# lane net with dropout
# lane_net = Lane_drop()

# send to gpu
lane_net = lane_net.to(device=device)

# PATH = 'dense_lane.pt'
# lane_net.load_state_dict(torch.load(PATH))
# lane_net.eval()
# optimizer = optim.SGD(draw_net.parameters(),lr=0.001, momentum=0.9)
optimizer = optim.Adam(lane_net.parameters(),lr=0.001)
criterion = nn.CrossEntropyLoss()
EPOCHS = 100
```

```python
def test_acc():
    predictions = []
    gts = []
    for batch_idx, (X_batch, y_batch) in enumerate(test_loader):
        var_X_batch = Variable(X_batch).float().to(device=device)
        var_y_batch = Variable(y_batch).to(device=device)

        output = lane_net(var_X_batch)
        predicted = torch.max(output.data, 1)[1] 

        predictions.append(predicted.cpu().detach().numpy()[0])
        gts.append(var_y_batch.cpu().detach().numpy()[0])
    total_train_accuracy = (np.array(predictions) == np.array(gts)).sum() / len(predictions) * 100 
    return total_train_accuracy
```
